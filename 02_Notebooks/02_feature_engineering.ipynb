# 02- Feature Engineering

## Objectif Bloc 5
Construction des variables exploitables pour le modèle.

```python
import pandas as pd
import numpy as np

df = pd.read_csv("../01_Data/raw/data.csv")

df["TotalCharges"] = pd.to_numeric(df["TotalCharges"], errors="coerce")
df["TotalCharges"] = df["TotalCharges"].fillna(df["TotalCharges"].median())
df["Churn"] = df["Churn"].map({"No": 0, "Yes": 1})

df.head()
```

## Interprétation
La cible est transformée en binaire et TotalCharges est convertie en numérique.


```python
df["avg_charge_per_tenure"] = df["MonthlyCharges"] / df["tenure"].replace(0, 1)
df["has_fiber"] = (df["InternetService"] == "Fiber optic").astype(int)
df["is_month_to_month"] = (df["Contract"] == "Month-to-month").astype(int)
df["has_online_security"] = (df["OnlineSecurity"] == "Yes").astype(int)
df["has_tech_support"] = (df["TechSupport"] == "Yes").astype(int)

df[[
    "avg_charge_per_tenure",
    "has_fiber",
    "is_month_to_month",
    "has_online_security",
    "has_tech_support",
    "Churn"
]].head()
```

## Interprétation
Ces variables traduisent des facteurs métier : engagement, service, assistance, sécurité et facturation.


```python
df.to_csv("../01_Data/processed/churn_features.csv", index=False)
print("Dataset enrichi sauvegardé.")
```
